```python
import pandas as pd
import numpy as np
```

```python
matches = pd.read_csv("../data/matches.csv")
deliveries = pd.read_csv("../data/deliveries.csv")
```

```python
matches['date'] = pd.to_datetime(matches['date'], errors='coerce')
```

```python
matches.drop_duplicates(inplace=True)
deliveries.drop_duplicates(inplace=True)
```

```python
cols_to_clean = ['team1', 'team2', 'winner', 'venue']
for col in cols_to_clean:
    matches[col] = matches[col].astype(str).str.strip()

deliveries['batting_team'] = deliveries['batting_team'].astype(str).str.strip()
deliveries['bowling_team'] = deliveries['bowling_team'].astype(str).str.strip()
```

```python
print(matches.isna().sum())
print(deliveries.isna().sum())
```

```python
matches.to_csv("../data/matches_clean.csv", index=False)
deliveries.to_csv("../data/deliveries_clean.csv", index=False)
```
