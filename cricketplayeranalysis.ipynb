```python
import pandas as pd

# Load ball-by-ball dataset
df = pd.read_csv("dataset/deliveries_updated_ipl_upto_2025.csv")

# ---------- BASIC CLEANUP ----------
# Standard column names (depends on your dataset)
df.columns = df.columns.str.lower()

# Runs scored per ball
df['total_runs'] = df['batsman_runs'] + df.get('extras', 0)

# Filter legal deliveries for strike rate & economy
df_legal = df[df['ball'].notna()]
```

```python
top_runs = df.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(15)
print("ðŸ† Top Run Scorers:")
print(top_runs)
```

```python
balls_faced = df.groupby('batsman').size()
runs_scored = df.groupby('batsman')['batsman_runs'].sum()

strike_rate = (runs_scored / balls_faced * 100).dropna()
strike_rate = strike_rate[balls_faced > 200].sort_values(ascending=False).head(15)

print("\nâš¡ Top Strike Rates (min 200 balls):")
print(strike_rate)
```

```python
sixes = df[df['batsman_runs'] == 6].groupby('batsman').size().sort_values(ascending=False).head(15)
fours = df[df['batsman_runs'] == 4].groupby('batsman').size().sort_values(ascending=False).head(15)

print("\nðŸ’¥ Most Sixes:")
print(sixes)

print("\nðŸ”¥ Most Fours:")
print(fours)
```

```python
df_wickets = df[df['dismissal_kind'].notnull()]
top_wickets = df_wickets.groupby('bowler').size().sort_values(ascending=False).head(15)

print("\nðŸŽ¯ Top Wicket Takers:")
print(top_wickets)
```

```python
runs_given = df.groupby('bowler')['total_runs'].sum()
balls_bowled = df[df['extras'] != 'wides'].groupby('bowler').size()

eco = (runs_given / (balls_bowled / 6)).dropna()
eco = eco[balls_bowled > 300].sort_values().head(15)

print("\nðŸ›¡ Best Economy Rates (min 300 balls):")
print(eco)
```

```python
bowler_sr = (balls_bowled / top_wickets).dropna()
bowler_sr = bowler_sr[balls_bowled > 300].sort_values().head(15)

print("\nðŸ”¥ Best Bowler Strike Rates (min 300 balls):")
print(bowler_sr)
```

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print("Starting IPL Data Cleaning...")

# =====================================================
# 1) Create Export Directory
# =====================================================
EXPORT_DIR = "/content/exports"
os.makedirs(EXPORT_DIR, exist_ok=True)
print("Export directory ready:", EXPORT_DIR)

# =====================================================
# 2) Load Dataset
# =====================================================
df = pd.read_csv("/content/dataset/deliveries_updated_ipl_upto_2025.csv")
print("Dataset loaded successfully!")
print("Rows:", len(df))
print("Columns:", df.columns.tolist())

# =====================================================
# 3) Fix dtypes
# =====================================================
df["date"] = pd.to_datetime(df["date"], errors="coerce")

numeric_cols = ["batsman_runs","extras","Byes","LegByes","Penalty"]
for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce").fillna(0)

# Convert wides/noballs to boolean
df["isWide"] = df["isWide"].fillna(0).astype(int)
df["isNoBall"] = df["isNoBall"].fillna(0).astype(int)

# =====================================================
# 4) Create "total_runs" Column
# =====================================================
df["total_runs"] = (
    df["batsman_runs"]
    + df["extras"]
    + df["Byes"]
    + df["LegByes"]
    + df["Penalty"]
)

# =====================================================
# 5) Handle Missing Values
# =====================================================
df["batsman"].fillna("Unknown", inplace=True)
df["bowler"].fillna("Unknown", inplace=True)
df["batting_team"].fillna("Unknown", inplace=True)
df["bowling_team"].fillna("Unknown", inplace=True)

# Missing dismissal fields
df["dismissal_kind"] = df["dismissal_kind"].fillna("Not Dismissed")
df["player_dismissed"] = df["player_dismissed"].fillna("None")

# =====================================================
# 6) Remove Invalid Balls (Outliers)
# =====================================================
df = df[df["ball"] >= 1]            # valid ball numbers
df = df[df["ball"] <= 6]            # max 6 balls per over
df = df[df["inning"].isin([1,2])]   # remove super-over inconsistencies

# =====================================================
# 7) Logical Corrections
# =====================================================
# If it's a wide or noball, batsman_runs should not exceed 6
df.loc[df["isWide"] == 1, "batsman_runs"] = df.loc[df["isWide"] == 1, "batsman_runs"].clip(0,6)
df.loc[df["isNoBall"] == 1, "batsman_runs"] = df.loc[df["isNoBall"] == 1, "batsman_runs"].clip(0,6)

# =====================================================
# 8) Outlier Treatment for Total Runs
# =====================================================
Q1 = df["total_runs"].quantile(0.25)
Q3 = df["total_runs"].quantile(0.75)
IQR = Q3 - Q1

lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

df["total_runs_capped"] = df["total_runs"].clip(lower_limit, upper_limit)

# =====================================================
# 9) Visual Plots Export
# =====================================================
plt.figure(figsize=(10,4))
sns.boxplot(x=df["total_runs"])
plt.title("Boxplot â€“ Total Runs")
plt.savefig(f"{EXPORT_DIR}/boxplot_total_runs.png")
plt.close()

plt.figure(figsize=(10,4))
sns.histplot(df["total_runs"], bins=30, kde=True)
plt.title("Histogram â€“ Total Runs")
plt.savefig(f"{EXPORT_DIR}/hist_total_runs.png")
plt.close()

# =====================================================
# 10) Save Cleaned Dataset
# =====================================================
clean_path = f"{EXPORT_DIR}/ipl_deliveries_cleaned.csv"
df.to_csv(clean_path, index=False)

print("âœ” CLEANED FILE SAVED:", clean_path)

print("\nðŸŽ‰ IPL CLEANING COMPLETE!")
print(f"âœ” ALL EXPORTS SAVED IN: {EXPORT_DIR}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import json
import os

# Create export folder
os.makedirs("exports", exist_ok=True)

# Load dataset
df = pd.read_csv("/content/dataset/deliveries_updated_ipl_upto_2025.csv")
print("Dataset loaded successfully!\n")

# -----------------------------------------------------------
# 1ï¸âƒ£ Fix missing fields manually
# -----------------------------------------------------------

# total_runs = batsman_runs + extras
df["total_runs"] = df["batsman_runs"] + df["Byes"] + df["LegByes"] + df["Penalty"] + df["extras"]

# wide balls exclude from ball counting
df["legal_ball"] = df.apply(lambda row: 0 if row["isWide"] == 1 or row["isNoBall"] == 1 else 1, axis=1)

# wicket flag
df["is_wicket"] = df["player_dismissed"].notna().astype(int)

# -----------------------------------------------------------
# 2ï¸âƒ£ TOP RUN SCORERS
# -----------------------------------------------------------

top_runs = df.groupby("batsman")["batsman_runs"].sum().sort_values(ascending=False)
top_runs.to_csv("exports/top_run_scorers.csv")

# -----------------------------------------------------------
# 3ï¸âƒ£ STRIKE RATE
# -----------------------------------------------------------

balls_faced = df[df["legal_ball"] == 1].groupby("batsman")["legal_ball"].count()
strike_rates = (df.groupby("batsman")["batsman_runs"].sum() / balls_faced * 100).dropna()
strike_rates.to_csv("exports/strike_rates.csv")

# -----------------------------------------------------------
# 4ï¸âƒ£ SIXES & FOURS
# -----------------------------------------------------------

sixes = df[df["batsman_runs"] == 6].groupby("batsman")["batsman_runs"].count()
fours = df[df["batsman_runs"] == 4].groupby("batsman")["batsman_runs"].count()
sixes.to_csv("exports/most_sixes.csv")
fours.to_csv("exports/most_fours.csv")

# -----------------------------------------------------------
# 5ï¸âƒ£ WICKETS
# -----------------------------------------------------------

wickets = df[df["is_wicket"] == 1].groupby("bowler")["is_wicket"].count()
wickets.to_csv("exports/top_wicket_takers.csv")

# -----------------------------------------------------------
# 6ï¸âƒ£ ECONOMY RATE
# -----------------------------------------------------------

balls_bowled = df[df["legal_ball"] == 1].groupby("bowler")["legal_ball"].count()
runs_conceded = df.groupby("bowler")["total_runs"].sum()

economy = (runs_conceded / (balls_bowled / 6)).replace([float("inf"), -float("inf")], 0)
economy.to_csv("exports/economy_rates.csv")

# -----------------------------------------------------------
# 7ï¸âƒ£ BOWLING STRIKE RATE (balls per wicket)
# -----------------------------------------------------------

bowling_strike_rate = (balls_bowled / wickets).dropna()
bowling_strike_rate.to_csv("exports/bowling_strike_rate.csv")

# -----------------------------------------------------------
# 8ï¸âƒ£ PLAYER JSON DATA
# -----------------------------------------------------------

player_stats = df.groupby("batsman")["batsman_runs"].sum().reset_index()
player_stats.to_json("exports/player_stats.json", orient="records")

# -----------------------------------------------------------
# 9ï¸âƒ£ PLOTS
# -----------------------------------------------------------

plt.figure(figsize=(10,5))
top_runs.head(10).plot(kind="bar")
plt.title("Top 10 Run Scorers")
plt.savefig("exports/top_scorers_plot.png")
plt.close()

plt.figure(figsize=(10,5))
wickets.head(10).plot(kind="bar", color="red")
plt.title("Top 10 Wicket Takers")
plt.savefig("exports/top_wickets_plot.png")
plt.close()

print("âœ” ALL FILES SAVED IN: /content/exports")
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Load data
df = pd.read_csv("compressed_data.csv")

# Select numeric columns only
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Two scalers
minmax = MinMaxScaler()
standard = StandardScaler()

# Apply MinMax Scaling
df_minmax = df.copy()
df_minmax[numeric_cols] = minmax.fit_transform(df[numeric_cols])
df_minmax.to_csv("normalized_minmax.csv", index=False)

# Apply Standard Scaling
df_standard = df.copy()
df_standard[numeric_cols] = standard.fit_transform(df[numeric_cols])
df_standard.to_csv("normalized_standard.csv", index=False)

print("Normalization complete!")
print("âœ” Saved: normalized_minmax.csv")
print("âœ” Saved: normalized_standard.csv")
```

```python
# ===========================
# IPL EDA COMPLETE SCRIPT
# ===========================

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

print("Loading data...")

df = pd.read_csv("compressed_data.csv")
print("Dataset loaded!")

# ===========================
# Create exports folder
# ===========================
export_dir = "/content/ipl_eda_exports"
os.makedirs(export_dir, exist_ok=True)
print(f"âœ” Export folder created at: {export_dir}")

# ===========================
# Fix for delivery-level wickets
# ===========================
df["is_wicket"] = df["dismissal_kind"].notna().astype(int)

# ===========================
# 1ï¸âƒ£ DISTRIBUTION OF RUNS
# ===========================

plt.figure(figsize=(8,5))
sns.histplot(df["batsman_runs"], kde=True)
plt.title("Distribution of Per-ball Batsman Runs")
plt.xlabel("Runs per ball")
plt.ylabel("Frequency")
plt.savefig(f"{export_dir}/distribution_batsman_runs.png")
plt.close()

# Summary table
runs_summary = df["batsman_runs"].describe()
runs_summary.to_csv(f"{export_dir}/runs_distribution_summary.csv")

print("âœ” Runs distribution done.")

# ===========================
# 2ï¸âƒ£ DISTRIBUTION OF WICKETS
# ===========================

plt.figure(figsize=(7,5))
sns.countplot(x=df["is_wicket"])
plt.title("Distribution of Wickets per Ball")
plt.xlabel("Wicket (0=No, 1=Yes)")
plt.ylabel("Count")
plt.savefig(f"{export_dir}/distribution_wickets.png")
plt.close()

wickets_rate = df["is_wicket"].mean()
pd.DataFrame({"wicket_rate_per_ball": [wickets_rate]}).to_csv(
    f"{export_dir}/wicket_rate.csv", index=False
)

print("âœ” Wickets distribution done.")

# ===========================
# 3ï¸âƒ£ VENUE STATISTICS (Requires matches file)
# ===========================

print("Loading matches file for venue stats...")

# Try all possible filenames
matches_files = [
    "matches_updated_ipl_upto_2025.csv",
    "matches_updated_mens_ipl.csv",
    "matches.csv",
]

matches_path = None
for f in matches_files:
    if os.path.exists(f):
        matches_path = f
        break

if matches_path:
    matches = pd.read_csv(matches_path)
    print(f"âœ” Matches file found: {matches_path}")

    # Top venues by matches
    top_venues = matches["venue"].value_counts().head(10)
    top_venues.to_csv(f"{export_dir}/top_venues.csv")

    plt.figure(figsize=(10,5))
    sns.barplot(x=top_venues.values, y=top_venues.index)
    plt.title("Top IPL Venues by Number of Matches")
    plt.xlabel("Matches Played")
    plt.savefig(f"{export_dir}/top_venues.png")
    plt.close()

    print("âœ” Venue stats done.")
else:
    print("âš  No matches file found. Venue stats skipped.")

# ===========================
# 4ï¸âƒ£ TEAM PERFORMANCE
# ===========================

print("Calculating team performance...")

team_runs = df.groupby("batting_team")["batsman_runs"].sum().sort_values(ascending=False)
team_runs.to_csv(f"{export_dir}/team_total_runs.csv")

plt.figure(figsize=(10,6))
sns.barplot(x=team_runs.values, y=team_runs.index)
plt.title("Total Runs Scored by Each IPL Team")
plt.xlabel("Total Runs")
plt.savefig(f"{export_dir}/team_total_runs.png")
plt.close()

# Wickets taken per team
team_wickets = df.groupby("bowling_team")["is_wicket"].sum().sort_values(ascending=False)
team_wickets.to_csv(f"{export_dir}/team_total_wickets.csv")

plt.figure(figsize=(10,6))
sns.barplot(x=team_wickets.values, y=team_wickets.index)
plt.title("Total Wickets Taken by Each Team")
plt.xlabel("Total Wickets")
plt.savefig(f"{export_dir}/team_total_wickets.png")
plt.close()

print("âœ” Team performance stats done.")

# ===========================
# FINISH
# ===========================
print("\nðŸŽ‰ EDA Completed Successfully!")
print(f"ðŸ“ All results saved in: {export_dir}")
```
